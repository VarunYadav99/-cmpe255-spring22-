```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import seaborn as sns
```

```python
data.columns
```

```python
#Dropping stock and base salry bonus
data = data.drop(['basesalary','stockgrantvalue','bonus'],axis = 1)
```

```python
#**Dropping columns with less than 0.1 corr values**
data = data.drop(['rowNumber','Masters_Degree','Bachelors_Degree','Doctorate_Degree','Highschool', 'Some_College', 'Race_Asian', 'Race_White',
       'Race_Two_Or_More', 'Race_Black','Race_Hispanic', 'Race'], axis =1)
```

```python
data
```

```python
# dropping non num number values columns
data = data.drop(['company', 'level', 'title','location','tag','gender','otherdetails','Education'],axis = 1)
```

```python
plt.figure(figsize=(20,15))
sns.heatmap(data.corr(),annot = True,cmap ='RdPu')
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range = (0,1))
```

```python
data.isnull().sum()
```

```python
data2 = data.dropna()
```

```python
data2.isnull().sum()
```

```python
data2
```

```python
Y = data2['totalyearlycompensation']
```

```python
X = data2.drop('totalyearlycompensation',axis =1)
X
```

```python
scaler.fit(X)
```

```python
 # tranforming X in range b/w 0,1
X_scaler = scaler.transform(X)
    
```

```python
X_scaler
```

```python
#Split into train and test
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test, = train_test_split(X_scaler, Y, test_size = 0.2, random_state =9)
```

```python
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression(normalize = True)
```

```python
#Training model with train data set
linear_model.fit(X_train, Y_train)
```

```python
linear_model.score(X_test, Y_test)
```

```python
yhat = linear_model.predict(X_test)
```

```python
import numpy as np
```

```python
#RMSE
from sklearn.metrics import mean_squared_error
np.sqrt(mean_squared_error(Y_test,yhat))
```
